Clssification

```python
import pandas as pd
import numpy as np
from mlxtend.feature_selection import ExhaustiveFeatureSelector as efs
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("iris.csv")
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1], test_size=0.3, stratify=data.iloc[:,-1], random_state=42)
```

```python
print(np.mean(cross_val_score(LogisticRegression(max_iter=1000), X_train, y_train, cv=5)))
print(np.mean(cross_val_score(LogisticRegression(max_iter=1000), X_test, y_test, cv=5)))
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
print("Accuracy before feature selection:", accuracy_score(y_test, y_pred))
```

```python
sel = efs(LogisticRegression(max_iter=1000), max_features=4,scoring='accuracy', cv=5)

sel.fit(X_train, y_train)
```

```python
sel.best_feature_names_
```

```python
sel.best_score_
```

```python
sel.subsets_
```

```python
colnames = X_train.columns[list(sel.best_idx_)]
print(colnames)
X_train_new = X_train[colnames]
X_test_new = X_test[colnames]
X_test_new.head()
```

```python
print(np.mean(cross_val_score(LogisticRegression(max_iter=1000),X_train_new, y_train, cv=5)))
print(np.mean(cross_val_score(LogisticRegression(max_iter=1000),X_test_new, y_test, cv=5)))
```

```python

model.fit(X_train_new, y_train)
y_pred = model.predict(X_test_new)
accuracy_score(y_test, y_pred)
print("Accuracy after feature selection:", accuracy_score(y_test, y_pred))
```

Regression

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
X = df.drop('medv', axis=1)
y = df['medv']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(r2_score(y_test, y_pred))
```

```python
sel2 = efs(LinearRegression(), max_features=13, scoring='r2', cv=5)
sel2.fit(X_train, y_train)
```

```python
sel2.best_feature_names_
```

```python
sel2.best_score_
```

```python
X_train_new = sel2.transform(X_train)
X_test_new = sel2.transform(X_test)
```

```python
model.fit(X_train_new, y_train)
y_pred = model.predict(X_test_new)
print(r2_score(y_test, y_pred))
```

Sequential Backward Elimination


```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
sbs = sfs(LinearRegression(), k_features="best", forward=False, scoring='r2', cv=5)
sbs.fit(X_train, y_train)
sbs.k_feature_names_
```

```python
x_train_new = sbs.transform(X_train)
x_test_new = sbs.transform(X_test)
model.fit(x_train_new, y_train)
y_pred = model.predict(x_test_new)
print(r2_score(y_test, y_pred))
```

Sequential Forward Elimination

```python
sfs = sfs(LinearRegression(), k_features="best", forward=True, scoring='r2', cv=5)
sfs.fit(X_train, y_train)
sfs.k_feature_names_
```
