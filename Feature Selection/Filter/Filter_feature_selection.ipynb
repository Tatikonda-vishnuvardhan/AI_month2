*Before Feature Selection*

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
le1 = LabelEncoder()
y1 = le1.fit_transform(y1)
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)
model = LogisticRegression(max_iter=1000)
model.fit(X1_train, y1_train)
y1_pred = model.predict(X1_test)
accuracy = accuracy_score(y1_test, y1_pred)
print(f"Accuracy before feature selection: {accuracy:.4f}")
```

**Feature Selection**

Duplicates

```python
X2 = data.drop(columns=["Activity"])
y2 = data["Activity"]
```

```python
X2.head()
```

```python
X2.T.duplicated().sum()
```

```python

X2 = X2.loc[:, ~X2.T.duplicated()]
```

```python
X2.head()
```

Variance Threshold

```python
var = X2.var()
print(var.sort_values(ascending=False))
```

```python
from sklearn.feature_selection import VarianceThreshold

# Remove features with low variance
selector = VarianceThreshold(threshold=0.05)
X2_red = selector.fit_transform(X2)
X2= pd.DataFrame(X2_red, columns=X2.columns[selector.get_support()])
X2.head()
```

```python
le2 = LabelEncoder()
y2 = le2.fit_transform(y2)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)
```

Correlation

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
sns.heatmap(X2_train.corr(), annot=False, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
corr_matrix = X2_train.corr()
corr_matrix
```

```python
columns = corr_matrix.columns
to_drop = []

for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        if corr_matrix.iloc[i, j] > 0.96 or corr_matrix.iloc[i, j] < -0.96:
            to_drop.append(columns[j])

to_drop = set(to_drop)
print(f"Columns to drop due to high correlation: {len(to_drop)}")
print(list(to_drop))

# Drop from both train and test
X2_train = X2_train.drop(columns=to_drop)
X2_test = X2_test.drop(columns=to_drop)
```

```python
print(f"Shape after removing correlated features: {X2_train.shape}")
print(f"Shape after removing correlated features: {X2_test.shape}")
```

Anova

```python
from sklearn.feature_selection import SelectKBest, f_classif
f_values, p_values = f_classif(X2_train, y2_train)
selected_features = X2_train.columns[p_values < 0.05]
print(f"Number of features selected by ANOVA: {len(selected_features)}")
```

```python
selector = SelectKBest(score_func=f_classif, k=150)
X2_train_selected = selector.fit_transform(X2_train, y2_train)
X2_test_selected = selector.transform(X2_test)
columns_selected = X2_train.columns[selector.get_support()]
X2_train = pd.DataFrame(X2_train_selected, columns=columns_selected)
X2_test = pd.DataFrame(X2_test_selected, columns=columns_selected)
```

Mutual Information

```python
from sklearn.feature_selection import mutual_info_classif
selctor = SelectKBest(score_func=mutual_info_classif, k=120)
X2_train_selected = selctor.fit_transform(X2_train, y2_train)
X2_test_selected = selctor.transform(X2_test)
columns_selected = X2_train.columns[selctor.get_support()]
X2_train = pd.DataFrame(X2_train_selected, columns=columns_selected)
X2_test = pd.DataFrame(X2_test_selected, columns=columns_selected)
```

```python
model.fit(X2_train, y2_train)
y2_pred = model.predict(X2_test)
accuracy2 = accuracy_score(y2_test, y2_pred)
print(f"Accuracy after feature selection: {accuracy2:.4f}")
```
