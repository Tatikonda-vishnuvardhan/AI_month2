```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("salary_data.csv")
df
```

```python
x= df[["YearsExperience"]]
y= df["Salary"]
```

```python
x
```

```python
y
```

```python
%matplotlib inline
plt.scatter(x,y,s=100,c="green",marker="+",alpha=1,label = "Scatter Plot")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("YearsExperience vs Salary")
plt.legend()
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
y_test
```

```python
mean_absolute_error(y_pred,y_test)
```

```python
mean_squared_error(y_pred,y_test)
```

```python
r2_score(y_pred,y_test)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].scatter(x_train, y_train, color='blue', label='Train Data')
axes[0].scatter(x_test, y_test, color='red', label='Test Data')
axes[0].set_xlabel("Years of Experience")
axes[0].set_ylabel("Salary")
axes[0].set_title("Linear Regression")
axes[0].legend()



axes[1].scatter(x_train, y_train, color='blue', label='Train Data')
axes[1].scatter(x_test, y_pred, color='red', label='Pred Data')
axes[1].set_xlabel("Years of Experience")
axes[1].set_ylabel("Salary")
axes[1].set_title("Linear Regression")
axes[1].legend()

plt.show()
```

```python
plt.scatter(x_test, y_test, color='blue', label='Actual Test Data',alpha=0.4)
plt.scatter(x_test, y_pred, color='red', label='Predicted Test Data',alpha=0.4)
plt.plot(x_test,y_pred,color='red', label='Predicted Test Data',)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression: Salary Predictor")
plt.legend()
plt.show()
```
