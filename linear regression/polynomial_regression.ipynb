```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler , PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x = 6 * np.random.rand(200, 1) - 3
y = 0.8 * x**2 + 0.9 * x + 2 + np.random.randn(200, 1)

# y = 0.8x^2 + 0.9x + 2
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,label = "generated data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

linear fit

```python
x_trian, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
lr = LinearRegression()
```

```python
lr.fit(x_trian,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
r2_score(y_pred,y_test)
```

```python
plt.scatter(x,y,label = "generated data")
plt.plot(x_test,lr.predict(x_test), "r")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

poly fit

degree = 2

```python
poly1 = PolynomialFeatures(degree=2)
x_train_trans1 = poly1.fit_transform(x_trian)
x_test_trans1 = poly1.transform(x_test)
```

```python
plr1 = LinearRegression()
plr1.fit(x_train_trans1,y_train)
```

```python
py_pred1 = plr1.predict(x_test_trans1)
```

```python
plr1.coef_
```

```python
plr1.intercept_
```

```python
r2_score(py_pred1,y_test)
```

```python
x_1 = np.linspace(-3,3,200).reshape(200,1)
x_1_poly = poly1.transform(x_1)
y_1 = plr1.predict(x_1_poly)
```

```python
plt.scatter(x,y,label = "generated data")
plt.plot(x_test,py_pred1,"r")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y,label = "generated data")
plt.plot(x_1,y_1,"r")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
