```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
import seaborn as sns
```

```python
data, true_labels = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)
```

```python
data
```

```python
plt.scatter(data[:, 0], data[:, 1], s=30)
plt.title("Generated Data")
plt.show()
```

```python
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(data)
```

```python
print("Means:\n", gmm.means_)
```

```python
predicted_labels = gmm.predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=predicted_labels, s=30, cmap='viridis')
plt.title("GMM Clustering Results")
plt.show()
```

```python
new_data = np.array([[0, 0], [10, 10], [-10, -10]])
predictions = gmm.predict(new_data)
pred_probabilities = gmm.predict_proba(new_data)
print("Predictions for new data points:\n", predictions)
print("Prediction probabilities for new data points:\n", pred_probabilities)
```
