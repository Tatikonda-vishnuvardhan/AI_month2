```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("imbalanced_dataset.csv")
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], test_size=0.2, random_state=42, stratify=data.iloc[:, -1])
```

```python
X_train.shape
```

```python
rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_train, y_train)
```

```python
X_resampled.shape
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_train.iloc[:, 0], X_train.iloc[:, 1], c=y_train,cmap='coolwarm')
class_labels = y_train.unique()
cmap = plt.get_cmap('coolwarm',len(class_labels))
for label in class_labels:
    color = cmap(label)
    plt.scatter([], [], c=color, label=f'Class {label}')
plt.legend(title = 'labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Original Imbalanced Data')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(X_resampled.iloc[:, 0], X_resampled.iloc[:, 1], c=y_resampled,cmap='coolwarm',label='class 0')
plt.scatter([], [],color ="red", label='Class 1')
plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('After Random Under-Sampling')
plt.show()
```
