```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.head()
```

UNIVARIATE

```python
sns.kdeplot(cars.horsepower, fill=True)
```

```python
sns.kdeplot(cars.horsepower, fill=True,bw_adjust=0.5)
```

BIVARIATE

```python
sns.kdeplot(x=cars.horsepower,y= cars.mpg)
```

```python
sns.kdeplot(x=cars.horsepower,y= cars.mpg,levels=20)
```

```python
sns.kdeplot(x=cars.horsepower,y= cars.mpg, fill=True, cmap='Reds' )
```
